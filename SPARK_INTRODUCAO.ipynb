```python

def get_file_paths(dir_name):
    dir_path = os.path.abspath(dir_name)
    res = []
    for path in os.listdir(dir_path):
        if os.path.isfile(os.path.join(dir_path, path)):
            res.append(os.path.join(dir_path, path))
    return res

# Obtem os caminhos dos arquivos para cada diretório
estabelecimentos_files = get_file_paths(path_estabelicimentos)
empresas_files = get_file_paths(path_empresas)
socios_files = get_file_paths(path_socios)

# Lê os arquivos em DataFrames Spark
estabelecimentos_df = spark.read.csv(estabelecimentos_files, sep=';', inferSchema=True)
empresas_df = spark.read.csv(empresas_files, sep=';', inferSchema=True)
socios_df = spark.read.csv(socios_files, sep=';', inferSchema=True)
```

# CONVERTENDO DATAS

usar a columns data_de_entrada_sociedade do schema socios para convereter para date atraves da function Date.


```python
socios = socios.withColumn('data_de_entrada_sociedade', 
                           f.to_date(socios['data_de_entrada_sociedade'].cast(StringType()), 'yyyMMdd'))

socios.limit(5).toPandas()
```

### Converter dados da table Estabelecimentos para Date

root
 |-- data_situacao_cadastral: integer (nullable = true)
 |-- data_de_inicio_atividade: integer (nullable = true)
 |-- data_da_situacao_especial: integer (nullable = true)

```python
columns_date = ['data_situacao_cadastral', 'data_de_inicio_atividade', 'data_da_situacao_especial']

for date in columns_date:
    estabelicimentos = estabelicimentos.withColumn(date, 
                                                   f.to_date(estabelicimentos[date].cast(StringType()),
                                                            'yyyMMdd'))
    
estabelicimentos.limit(10).toPandas()
```

# CRIAR SQL E BUSCA COM SQL

Primeiro precisamos criar uma view para trabalhar com SQL

```python
empresas.createOrReplaceTempView('empresasView')
```

```python
spark.sql(
"""
    SELECT * FROM empresasView
"""
).limit(5).toPandas()
```

```python
spark.sql(
"""
    SELECT * 
        FROM empresasView
        WHERE capital_social_da_empresa = 50
"""
).limit(5).toPandas()
```

```python
spark.sql(
"""
    SELECT porte_da_empresa, MEAN(capital_social_da_empresa) AS media 
        FROM empresasView
        GROUP BY porte_da_empresa
        
"""
).limit(5).toPandas()
```
